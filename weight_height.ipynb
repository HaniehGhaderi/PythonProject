```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from scipy.stats import poisson
from matplotlib import pyplot as plt
import seaborn as sns

ideal=norm.rvs(size=len(males),loc=males.describe()['mean'],scale=males.describe()['std'])
print(ideal)
#ساخت یک سری داده های توزیه شده ی نرمال به اندازه ی تعداد مرد هاو میانگین و انحراف معیار دیتاست اصلی
#قراره لیست اصلی با این دیتا ست های نرمال مقایسه بشه تا بشه فهمید که دیتاست اصلی نرمال هست یا خیر
```

```python
icount, idevision=np.histogram(ideal,bins=100)
print(icount)
#می گه از هر نمونه چندتاست
print(idevision)
#هیستوگرام توزیع نرمالی که ساختیم رو نشون میده
```

```python

def kl(p,q):
  result = np.sum(np.where(np.logical_and(p!=0,q!=0),p*np.log(p/q),0))
  return result

 #فرمولkl
 #با این تابع میشه به مقایسه دو توزیع اماری پرداخت
 #معیار واگرایی کولبک-لیبلر
```

```python
import numpy as np
import pandas as pd
import scipy.stats as st
from scipy.stats import norm
from scipy.stats import entropy
from scipy.stats import poisson


def Js(p,q):
  p1=p/np.linalg.norm(p,ord=1)
  q1=q/np.linalg.norm(q,ord=1)
  m=0/5*(p1+q1)
  r=0/5*(entropy(p1,m)+ entropy(q1,m))
  return r
#فرمولJs
#با این تابع میشه به مقایسه دو توزیع اماری پرداخت
#فاصله‌ی جنسون-شنون
```
